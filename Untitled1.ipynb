```python
data_cleaned = df.dropna(subset=['Salary', 'Year'])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data_cleaned['Year'], data_cleaned['Salary'], marker='o', linestyle='-', color='b', label='Salary')

plt.title('Salary Trend Over the Years ')
plt.xlabel('Year(X-Independent Variable)')
plt.ylabel('Salary(Y-Dependent Variable)')
plt.legend()

plt.show()
```

```python
data_cleaned['Year'] = data_cleaned['Year'].astype(int)
data_cleaned['Salary'] = data_cleaned['Salary'].astype(float)

avg_salary_per_year = data_cleaned.groupby('Year')['Salary'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(avg_salary_per_year['Year'], avg_salary_per_year['Salary'], color='b')

plt.title('Average Salary Trend Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Salary')
```

```python
data_cleaned = df.dropna(subset=['Experience Level', 'Salary'])
data_cleaned['Salary'] = data_cleaned['Salary'].astype(float)

avg_salary_per_experience = data_cleaned.groupby('Experience Level')['Salary'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(avg_salary_per_experience['Experience Level'], avg_salary_per_experience['Salary'], color='blue')

plt.title('Average Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary')
plt.show()
```
